```python
A = []

for xx in roadlst:
    for i in tqdm(range(len(xx))):        
        df999 = df3[df3['LINK_ID'] == xx[i]]
        df999 = df999.sort_values(by=['시간'])
        df999 = df999.reset_index(drop=True)
        df9999 = df999.groupby(['구분']).mean()

        result = pd.DataFrame(None)
        result['통행속도'] = df9999['통행속도']
        #real_result = result.T
        A.append(result.T)

# 20211107 경기도 상행 통행속도 합침
real_result = pd.concat(A)
real_result = real_result.reset_index(drop=True)

# 판다스 반올림
real_result = real_result.round(1)
real_result
```
