```python
import numpy as np
import pandas as pd

# (추가).. 데이터 용량 줄이기.. 일단은 필요한 링크들만 추출..
filtering = pd.read_excel('./사업지도로만_최종.xlsx')
filtering.head(3)

# filename = 20221107
filename = []

for xx in range(8): # 며칠할껀지
    filename.append(20221107 + xx)

for ss in range(len(filename)):
    df = pd.read_csv(f'./{filename[ss]}_5Min.csv')
    print(filename[ss],'중...')
    df.columns = ['날짜','시간','LINK_ID','도로기관코드','통행속도']
    df.head(3)

    df2 = df.copy()
    df2.head(3)

    ## 경기도

    df2['newone'] = df2['LINK_ID'].astype(str).str[:3]
    df3 = df2[df2['newone'] == '223']
    df3["구분"] = ((df3['시간']-1)/100).astype(int)
    df3
    
    HI = []

    for i in range(len(filtering)):
        HI.append(df3[df3['LINK_ID'] == filtering['LINK_ID'].values.tolist()[i]])

    filtering2 = pd.concat(HI)
    filtering2 = filtering2.reset_index(drop=True)
    filtering2

    filtering2.to_excel(f'./태블로식{filename[ss]}.xlsx')
    
print('FIN')
```
